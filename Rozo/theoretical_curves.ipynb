```python
import numpy as np
import matplotlib.pyplot as plt
from constants import RSOFT

plt.style.use(['science'])
```

```python
from matplotlib.lines import Line2D
from constants import PIVOT_MASS, RADIUS
from functions import rho_orb

fig, ax = plt.subplots(2, 3, sharex=True, sharey=False, figsize=(10,4), gridspec_kw={'height_ratios': [2.5, 1], 'hspace': 0.07}, dpi=300)

Morb = 10 ** 15
r_h = (843.8 / 1000) * (Morb / PIVOT_MASS)**(0.221)

rho = rho_orb(Morb, r_h)
ax[0, 0].plot(RADIUS, RADIUS**2 * rho, alpha=1, color='#ffd380', linewidth=0.8)
ax[0, 0].axvline(r_h, color='#ffd380', linewidth=0.8)

rho_1 = rho_orb(Morb, r_h*0.9)
ax[0, 0].plot(RADIUS, RADIUS**2 * rho_1, alpha=1, color='#ffd380', linewidth=0.8, linestyle='--')

rho_2 = rho_orb(Morb, r_h*1.1)
ax[0, 0].plot(RADIUS, RADIUS**2 * rho_2, alpha=1, color='#ffd380', linewidth=0.8, linestyle=':')

ax[1, 0].plot(RADIUS, rho/rho_1, alpha=1, color='#ffd380', linewidth=0.8, linestyle='--')
ax[1, 0].plot(RADIUS, rho/rho_2, alpha=1, color='#ffd380', linewidth=0.8, linestyle=':')


Morb = 10 ** 14
r_h = (843.8 / 1000) * (Morb / PIVOT_MASS)**(0.221)

h = np.linspace(r_h*0.9, r_h*1.1, 10)

rho = rho_orb(Morb, r_h)
ax[0, 0].plot(RADIUS, RADIUS**2 * rho, alpha=1, color='#ff6361', linewidth=0.8)
ax[0, 0].axvline(r_h, color='#ff6361', linewidth=0.8)


rho_1 = rho_orb(Morb, r_h*0.9)
ax[0, 0].plot(RADIUS, RADIUS**2 * rho_1, alpha=1, color='#ff6361', linewidth=0.8, linestyle='--')

rho_2 = rho_orb(Morb, r_h*1.1)
ax[0, 0].plot(RADIUS, RADIUS**2 * rho_2, alpha=1, color='#ff6361', linewidth=0.8, linestyle=':')

ax[1, 0].plot(RADIUS, rho/rho_1, alpha=1, color='#ff6361', linewidth=0.8, linestyle='--')
ax[1, 0].plot(RADIUS, rho/rho_2, alpha=1, color='#ff6361', linewidth=0.8, linestyle=':')

Morb = 10 ** 13.5
r_h = (843.8 / 1000) * (Morb / PIVOT_MASS)**(0.221)

h = np.linspace(r_h*0.9, r_h*1.1, 10)

rho = rho_orb(Morb, r_h)
ax[0, 0].plot(RADIUS, RADIUS**2 * rho, alpha=1, color='#003f5c', linewidth=0.8)
ax[0, 0].axvline(r_h, color='#003f5c', linewidth=0.8)

rho_1 = rho_orb(Morb, r_h*0.9)
ax[0, 0].plot(RADIUS, RADIUS**2 * rho_1, alpha=1, color='#003f5c', linewidth=0.8, linestyle='--')

rho_2 = rho_orb(Morb, r_h*1.1)
ax[0, 0].plot(RADIUS, RADIUS**2 * rho_2, alpha=1, color='#003f5c', linewidth=0.8, linestyle=':')

ax[1, 0].plot(RADIUS, rho/rho_1, alpha=1, color='#003f5c', linewidth=0.8, linestyle='--')
ax[1, 0].plot(RADIUS, rho/rho_2, alpha=1, color='#003f5c', linewidth=0.8, linestyle=':')

ax[0, 0].set_title(r'$r_h$')
ax[0, 0].set_xscale('log')
ax[0, 0].set_yscale('log')
# ax[1, 0].set_xscale('log')
# ax[1, 0].set_yscale('log')
ax[0, 0].set_xlim(1e-2, 10 ** 0.7)
ax[0, 0].set_ylim(1e9, 1e14)
ax[1, 0].set_ylim(0.8, 1.2)
ax[1, 0].set_xlabel(r'$r$ [Mpc/h]')
ax[0, 0].set_ylabel(r'$r^2 \rho_{orb}(r|M)$')
ax[1, 0].set_ylabel(r'ratio')

########################################

r_h = None
Morb = 10 ** 15
epsilon = 0.038

rho = rho_orb(Morb, r_h, epsilon)
ax[0, 1].plot(RADIUS, RADIUS**2 * rho, alpha=1, color='#ffd380', linewidth=0.8)

rho_1 = rho_orb(Morb, r_h, epsilon*0.9)
ax[0, 1].plot(RADIUS, RADIUS**2 * rho_1, alpha=1, color='#ffd380', linewidth=0.8, linestyle='--')

rho_2 = rho_orb(Morb, r_h, epsilon*1.1)
ax[0, 1].plot(RADIUS, RADIUS**2 * rho_2, alpha=1, color='#ffd380', linewidth=0.8, linestyle=':')

ax[1, 1].plot(RADIUS, rho/rho_1, alpha=1, color='#ffd380', linewidth=0.8, linestyle='--')
ax[1, 1].plot(RADIUS, rho/rho_2, alpha=1, color='#ffd380', linewidth=0.8, linestyle=':')


Morb = 10 ** 14
epsilon = 0.038

rho = rho_orb(Morb, r_h, epsilon)
ax[0, 1].plot(RADIUS, RADIUS**2 * rho, alpha=1, color='#ff6361', linewidth=0.8)

rho_1 = rho_orb(Morb, r_h, epsilon*0.9)
ax[0, 1].plot(RADIUS, RADIUS**2 * rho_1, alpha=1, color='#ff6361', linewidth=0.8, linestyle='--')

rho_2 = rho_orb(Morb, r_h, epsilon*1.1)
ax[0, 1].plot(RADIUS, RADIUS**2 * rho_2, alpha=1, color='#ff6361', linewidth=0.8, linestyle=':')

ax[1, 1].plot(RADIUS, rho/rho_1, alpha=1, color='#ff6361', linewidth=0.8, linestyle='--')
ax[1, 1].plot(RADIUS, rho/rho_2, alpha=1, color='#ff6361', linewidth=0.8, linestyle=':')

Morb = 10 ** 13.5
epsilon = 0.038

rho = rho_orb(Morb, r_h, epsilon)
ax[0, 1].plot(RADIUS, RADIUS**2 * rho, alpha=1, color='#003f5c', linewidth=0.8)

rho_1 = rho_orb(Morb, r_h, epsilon*0.9)
ax[0, 1].plot(RADIUS, RADIUS**2 * rho_1, alpha=1, color='#003f5c', linewidth=0.8, linestyle='--')

rho_2 = rho_orb(Morb, r_h, epsilon*1.1)
ax[0, 1].plot(RADIUS, RADIUS**2 * rho_2, alpha=1, color='#003f5c', linewidth=0.8, linestyle=':')

ax[1, 1].plot(RADIUS, rho/rho_1, alpha=1, color='#003f5c', linewidth=0.8, linestyle='--')
ax[1, 1].plot(RADIUS, rho/rho_2, alpha=1, color='#003f5c', linewidth=0.8, linestyle=':')

ax[0, 1].set_title(r'$\epsilon$')
ax[0, 1].set_xscale('log')
ax[0, 1].set_yscale('log')
ax[1, 1].set_xscale('log')
# ax[1, 1].set_yscale('log')
ax[0, 1].set_xlim(1e-2, 10 ** 0.7)
ax[0, 1].set_ylim(1e9, 1e14)
ax[1, 1].set_ylim(0.8, 1.2)
# ax[0, 1].set_xlabel(r'$r$ [Mpc/h]')
# ax[0, 1].set_ylabel(r'$r^2 \rho_{orb}(r|M)$')
# ax[1, 1].set_ylabel(r'ratio')

########################################

epsilon = None

Morb = 10 ** 15
alpha_inf = 2.021 * (Morb / PIVOT_MASS)**(-0.053)

rho = rho_orb(Morb, r_h, epsilon, alpha_inf)
ax[0, 2].plot(RADIUS, RADIUS**2 * rho, alpha=1, color='#ffd380', linewidth=0.8)

rho_1 = rho_orb(Morb, r_h, epsilon, alpha_inf*0.9)
ax[0, 2].plot(RADIUS, RADIUS**2 * rho_1, alpha=1, color='#ffd380', linewidth=0.8, linestyle='--')

rho_2 = rho_orb(Morb, r_h, epsilon, alpha_inf*1.1)
ax[0, 2].plot(RADIUS, RADIUS**2 * rho_2, alpha=1, color='#ffd380', linewidth=0.8, linestyle=':')

ax[1, 2].plot(RADIUS, rho/rho_1, alpha=1, color='#ffd380', linewidth=0.8, linestyle='--')
ax[1, 2].plot(RADIUS, rho/rho_2, alpha=1, color='#ffd380', linewidth=0.8, linestyle=':')

Morb = 10 ** 14
alpha_inf = 2.021 * (Morb / PIVOT_MASS)**(-0.053)

rho = rho_orb(Morb, r_h, epsilon, alpha_inf)
ax[0, 2].plot(RADIUS, RADIUS**2 * rho, alpha=1, color='#ff6361', linewidth=0.8)

rho_1 = rho_orb(Morb, r_h, epsilon, alpha_inf*0.9)
ax[0, 2].plot(RADIUS, RADIUS**2 * rho_1, alpha=1, color='#ff6361', linewidth=0.8, linestyle='--')

rho_2 = rho_orb(Morb, r_h, epsilon, alpha_inf*1.1)
ax[0, 2].plot(RADIUS, RADIUS**2 * rho_2, alpha=1, color='#ff6361', linewidth=0.8, linestyle=':')

ax[1, 2].plot(RADIUS, rho/rho_1, alpha=1, color='#ff6361', linewidth=0.8, linestyle='--')
ax[1, 2].plot(RADIUS, rho/rho_2, alpha=1, color='#ff6361', linewidth=0.8, linestyle=':')

Morb = 10 ** 13.5
alpha_inf = 2.021 * (Morb / PIVOT_MASS)**(-0.053)

rho = rho_orb(Morb, r_h, epsilon, alpha_inf)
ax[0, 2].plot(RADIUS, RADIUS**2 * rho, alpha=1, color='#003f5c', linewidth=0.8)

rho_1 = rho_orb(Morb, r_h, epsilon, alpha_inf*0.9)
ax[0, 2].plot(RADIUS, RADIUS**2 * rho_1, alpha=1, color='#003f5c', linewidth=0.8, linestyle='--')

rho_2 = rho_orb(Morb, r_h, epsilon, alpha_inf*1.1)
ax[0, 2].plot(RADIUS, RADIUS**2 * rho_2, alpha=1, color='#003f5c', linewidth=0.8, linestyle=':')

ax[1, 2].plot(RADIUS, rho/rho_1, alpha=1, color='#003f5c', linewidth=0.8, linestyle='--')
ax[1, 2].plot(RADIUS, rho/rho_2, alpha=1, color='#003f5c', linewidth=0.8, linestyle=':')

ax[0, 2].set_title(r'$\alpha_\infty$')
ax[0, 2].set_xscale('log')
ax[0, 2].set_yscale('log')
ax[1, 2].set_xscale('log')
# ax[1, 2].set_yscale('log')
ax[0, 2].set_xlim(1e-2, 10 ** 0.7)
ax[0, 2].set_ylim(1e9, 1e14)
ax[1, 2].set_ylim(0.8, 1.2)
# ax[0, 2].set_xlabel(r'$r$ [Mpc/h]')
# ax[0, 2].set_ylabel(r'$r^2 \rho_{orb}(r|M)$')
# ax[1, 2].set_ylabel(r'ratio')

# draw a horizonatal line at 1 for all the ratio plot
ax[1, 0].axhline(y=1, color='black', linestyle='-', linewidth=0.6, alpha=0.8)
ax[1, 1].axhline(y=1, color='black', linestyle='-', linewidth=0.6, alpha=0.8)
ax[1, 2].axhline(y=1, color='black', linestyle='-', linewidth=0.6, alpha=0.8)


# make a legend
legend_elements = [ Line2D([0], [0], color='#ffd380', lw=1, label=r'$M_{\rm orb} = 10^{15} M_\odot$'),
                    Line2D([0], [0], color='#ff6361', lw=1, label=r'$M_{\rm orb} = 10^{14} M_\odot$'),
                    Line2D([0], [0], color='#003f5c', lw=1, label=r'$M_{\rm orb} = 10^{13.5} M_\odot$'),
                    Line2D([0], [0], color='black', lw=1, linestyle='--', label=r'$-10\%$'),
                    Line2D([0], [0], color='black', lw=1, linestyle=':', label=r'$+10\%$'),
                    Line2D([0], [0], color='black', lw=1, linestyle='-', label=r'$0\%$')]
ax[0, 0].legend(handles=legend_elements[:3], loc='lower left', frameon=False)
ax[0, 1].legend(handles=legend_elements[3:], loc='lower left', frameon=False)

# plt.tight_layout()

# make gray region for all plots
ax[0, 0].axvspan(1e-2, 6 * RSOFT, alpha=0.6, color='grey')
ax[0, 1].axvspan(1e-2, 6 * RSOFT, alpha=0.6, color='grey')
ax[0, 2].axvspan(1e-2, 6 * RSOFT, alpha=0.6, color='grey')
ax[1, 0].axvspan(1e-2, 6 * RSOFT, alpha=0.6, color='grey')
ax[1, 1].axvspan(1e-2, 6 * RSOFT, alpha=0.6, color='grey')
ax[1, 2].axvspan(1e-2, 6 * RSOFT, alpha=0.6, color='grey')

plt.savefig('out/figures/rho_orb.jpg', bbox_inches='tight')
```
