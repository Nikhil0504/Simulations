```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use(["science"])  # type: ignore
# plt.rcParams.update({'figure.dpi': '200'})
plt.rcParams["text.usetex"] = True
plt.rcParams["figure.figsize"] = (10, 10)
plt.rcParams["xtick.major.size"] = 15
plt.rcParams["xtick.minor.size"] = 7
plt.rcParams["ytick.major.size"] = 15
plt.rcParams["ytick.minor.size"] = 7
plt.rcParams["xtick.labelsize"] = 15
plt.rcParams["ytick.labelsize"] = 15
plt.rcParams["font.size"] = 20
plt.rcParams["legend.fontsize"] = 20
```

```python
a = np.loadtxt('nohup.out')
mi = np.argmin(a[:, 1])
plt.scatter(a[:, 0], a[:, 1])
plt.axvline(a[:, 0][mi], label=r'$\sum \frac{|(\rho_{mod} - \rho_{data})|}{{(\epsilon \rho_{data})}}$')
# plt.axvline(a[:, 0][mi], label=r'$\sum \ln (1 + \frac{(\rho_{model} - \rho_{data})^2}{(\epsilon \rho_{data})^2})$')
plt.ylabel(r'$< \sigma_{jk} >$')
plt.xlabel(r'$\epsilon$')
plt.title(f'Min $\\epsilon$: {a[:, 1][mi]}')
plt.legend()
```
